```python
plt.figure(figsize=(13, 8))

plt.subplot(221)

plt.scatter(c['F8_obsC'], c['loggC'], s=10, c=c['TeffC'], cmap="rainbow_r", vmin=4400, vmax=10000)
plt.colorbar().set_label("Teff")
plt.xlabel("$F_8$")
plt.ylabel("$\log\;g$")
plt.xlim(0, .35)
plot_outline(newer_version=False)
plt.title("Bastien et al. (2013) Observed F8")

plt.subplot(222)

fitted_F8 = calc_F8(c['loggC'], c['TeffC'], c['MC'], S)
plt.scatter(fitted_F8, c['loggC'], s=10, c=c['TeffC'], cmap="rainbow_r", vmin=4400, vmax=10000)
plt.colorbar().set_label("Teff")
plt.xlabel("$F_8$")
plt.ylabel("$\log\;g$")
plt.xlim(0, .35)
plot_outline(newer_version=False)
plt.title("(B16) F8 w/ Individually-Fitted S")

plt.subplot(223)

s = 1 + np.zeros_like(c['loggC'])
f = calc_F8(c['loggC'], c['TeffC'], c['MC'], s)
plt.scatter(f, c['loggC'], s=10, c=c['TeffC'], cmap="rainbow_r", vmin=4400, vmax=10000)
plt.colorbar().set_label("Teff")
plt.xlabel("$F_8$")
plt.ylabel("$\log\;g$")
plt.xlim(0, .35)
plot_outline(newer_version=False)
plt.title("Model F8 with No Suppression")

plt.subplot(224)

s = calc_cranmer_S(c['TeffC'])
f = calc_F8(c['loggC'], c['TeffC'], c['MC'], s)
plt.scatter(f, c['loggC'], s=10, c=c['TeffC'], cmap="rainbow_r", vmin=4400, vmax=10000)
plt.colorbar().set_label("Teff")
plt.xlabel("$F_8$")
plt.ylabel("$\log\;g$")
plt.xlim(0, .35)
plot_outline(newer_version=False)
plt.title("Model F8 w/ Cranmer (2014) Suppression")

plt.tight_layout()
```
