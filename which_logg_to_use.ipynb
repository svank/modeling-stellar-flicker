```python
e_frac = catalog['e_F8logg'] / catalog['F8logg']
E_frac = catalog['E_F8logg'] / catalog['F8logg']
plt.hist(e_frac * 100, bins=70, range=(2, 12), label="F8-")
plt.hist(E_frac * 100, bins=70, range=(2, 12), label="F8+")

e_frac = catalog['e_loggH'] / catalog['loggH']
E_frac = catalog['E_loggH'] / catalog['loggH']
plt.hist(e_frac * 100, bins=70, range=(2, 12), label="H-")
plt.hist(E_frac * 100, bins=70, range=(2, 12), label="H+")

plt.xlabel("log$\,g$ Uncertainty (%)")
plt.ylabel("Count")
plt.title("$F_8$-derived log$\,g$ quotes lower uncertainties")
plt.legend();
```

```python
plt.hexbin(catalog['F8logg'], catalog['loggH'], vmax=30)
plt.xlabel("$F_8$-derived log$\,g$")
plt.ylabel("Huber log$\,g$")

plt.title("The log$\,g$ sources don't agree super well")
plt.colorbar().set_label("Count")

pearson_coeff, p_value = scipy.stats.pearsonr(catalog['F8logg'], catalog['loggH'])
print(f"Correlation coefficient r = {pearson_coeff:.2}")
```
